```python
df_train = df_train[df_train['Gender'] != '5']
print(df_train['Gender'].unique())
```

```python
df_train.info()
```

```python
df_train['Gender'] = df_train['Gender'].map({'Male': 0, 'Female': 1})
df_test['Gender'] = df_test['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df_train.info()
```

```python
numerical_columns = df_train.select_dtypes(include=['int64']).columns
```

```python
Q1 = df_train[numerical_columns].quantile(0.25)
Q3 = df_train[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df_train[numerical_columns] < (Q1 - 1.5 * IQR)) | (df_train[numerical_columns] > (Q3 + 1.5 * IQR)))
outliers_count = outliers.sum()

print("Columns with outliers:")
print(outliers_count[outliers_count > 0])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_columns = ['Age', 'openness', 'neuroticism', 'conscientiousness', 'agreeableness', 'extraversion']
df_train[num_columns] = scaler.fit_transform(df_train[num_columns])
```

```python
df_train.head()
```

```python
df_test[num_columns] = scaler.transform(df_test[num_columns])
```

```python
sns.pairplot(df_train, hue='Personality (Class label)')
plt.show()
```

```python
df_test.head()
```

```python
X_train, y_train=df_train.drop("Personality (Class label)",axis=1),df_train['Personality (Class label)']
```

```python
X_test, y_test=df_test.drop("Personality (class label)",axis=1),df_test['Personality (class label)']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
rf_model=RandomForestClassifier(random_state=42)
rf_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = rf_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
svc_model=SVC()
svc_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = svc_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
knc_model=KNeighborsClassifier()
knc_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = knc_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = lr_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
lr = LogisticRegression(multi_class='multinomial', solver='newton-cg',max_iter =1000)
lr.fit(X_train,y_train)
```

```python
lr_pred = lr.predict(X_test)
print(classification_report(y_test, lr_pred))
```
